```python
def get_text_len(text):
  return len(text)
```

```python
def get_weekday(unix_time):
    date_global = datetime.datetime.fromtimestamp(unix_time)
    return calendar.day_name[date_global.weekday()]
```

```python
def get_month(unix_time):
    date_month = datetime.datetime.fromtimestamp(unix_time)
    return calendar.month_name[date_month.month]
```

```python
def get_hour(unix_time):
    date_hour = datetime.datetime.fromtimestamp(unix_time)
    return datetime.datetime.hour[date_hour.hour]
```

```python
def switch_timestamp(unix_time):
    date_time = datetime.datetime.fromtimestamp(unix_time)
    return date_time
```

```python
def rel_mae(Y,P):
  Y=list(Y)
  P=list(P)
  S = [abs(y-p)/y for y,p in zip(Y,P)]
  return sum(S)/len(S)
```

```python
hist_boosting = HistGradientBoostingRegressor(loss='absolute_error')
hist_boosting.fit(X_scaled, Y)
y_predicted_test3 = hist_boosting.predict(X)
print(rel_mae(Y,y_predicted_test3))
```
