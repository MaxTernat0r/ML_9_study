### Задача 4

Кто получает больше: умные или красивые? (Придумайте сами, как это формализовать и опишите результат).

```python
# Для начала я бы хотел проверить два пункта: 1) Зависит ли индекс красоты от цвета кожи и здоровья работника, и 2) Зависит ли индекс красоты от пола работника
plt.scatter(data['looks'], data['black'])
plt.show()
plt.scatter(data['looks'], data['goodhlth'])
plt.show()
plt.scatter(data['looks'], data['female'])
plt.show()
```

```python
# Очевидно, почти никак не зависит. Я думаю, что это все факторы которые могли бы повлиять на красоту работника. Тогда сравним
# людей с восемью и меньше классами образования, но при этом индексом красоты 4 и выше, и людей с 11 и выше классами образования,
# но при этом индексом красоты 2 и ниже.

smart = data.loc[(data['looks'] <= 2)]
smart = smart.loc[(smart['educ'] >= 11)]
beautiful = data.loc[(data['looks'] >= 4)]
beautiful = beautiful.loc[(beautiful['educ'] <= 8)]

plt.hist(smart['wage'])
plt.hist(beautiful['wage'], color='red')
plt.show()
```

На данной гистограмме не стоит смотреть на высоту столбцов, так как мы не учитывали кол-во тех или иных работников, и данные
могут быть неправильными. Однако вы видим что зарплату больше 23 получают только умные, но некрасивые, и этого достаточно чтобы дать ответ.

Ответ: умные зарабатывают больше.

### Задача 5

Постройте линейную регрессию, предсказывающую красоту человека по остальным признакам. 

Вспомните, зачем нужно шаклировать признаки?

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import csv
```

```python
X = []
Y = []
total = 0
with open("beauty.csv", encoding='utf-8') as r_file:
    file_reader = csv.reader(r_file, delimiter=';')
    cnt = 0
    for row in file_reader:
        if cnt != 0:
            X.append(tuple([float(i) for i in row[0:9]]))
            Y.append(float(row[-1]))
        else:
            cnt = 1

model = LinearRegression()
model.fit(X, Y)
Y_predicted = model.predict(X)

S = [((y - y_predicted) / y) ** 2 for (y, y_predicted) in zip(Y, Y_predicted)]
relative_mse = (sum(S)/len(S))
print(f'Средний % ошибки {(relative_mse ** 0.5) * 100}, что достаточно много.')
```

# Задача 6
Интерпретируйте коэфициенты линейной регрессии: по каким признакам можно угадать, красивый ли человек?


```python
feature_names = ["wage", "exper", "union", "goodhlth", "black", "female", "married", "service", "educ"]

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

# Заново обучаем модель на уже масштабированных данных
model_scaled = LinearRegression()
model_scaled.fit(X_scaled, Y)

Y_predicted2 = model_scaled.predict(X_scaled)

print("Рассмотрим все коэффиценты датасета:")
for value, name in sorted(zip(model_scaled.coef_, feature_names), reverse=True):
    print(round(value, 2), "-", name)
print(
    "Получается, что educ влияет сильнее всего в положительную сторону, exper в отрицательную, а service не имеет влияния.")
```
